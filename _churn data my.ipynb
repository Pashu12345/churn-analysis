```python
sb.boxplot(A.numberVmailMessages)
```

```python
from scipy import stats
z=np.abs(stats.zscore(A.numberVmailMessages))
print(z)
```

We find the z-score for each of the data point in the dataset and if the z-score is greater than 3 than we can classify that point as an outlier. Any point outside of 3 standard deviations would be an outlier.

```python
threshold=3
print(np.where(z>3))
```

```python
A1=A[(z< 3)]
print(A1)
```

```python
X=A[["state","accountLength","areaCode","internationalPlan","voiceMailPlan","numberVmailMessages","totalDayCharge","totalDayMinutes","totalDayCalls","totalEveCalls","totalEveCharge","totalNightMinutes" , "totalNightCalls","totalNightCharge","totalIntlMinutes","totalIntlCalls","totalIntlCharge"]]
Y=A[["churner"]]

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=20)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(xtrain,ytrain)
pred=model.predict(xtest)
```

```python
ytest['predicted']=pred
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,auc
accuracy_score(ytest['churner'],pred)
#confusion_matrix(ytest['churner'],pred)
```

```python
pred=model.predict_proba(xtest)
fpr,tpr,thresholds=roc_curve(ytest['churner'],pred[:,1])
roc_auc=auc(fpr,tpr)
```

```python
import matplotlib.pyplot as plt
plt.plot(fpr,tpr,roc_auc)
```
